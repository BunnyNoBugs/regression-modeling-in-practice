```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
gm = pd.read_csv('gapminder.csv', na_values=' ', index_col=0)
gm
```

```python
gm.mean()
```

```python
for i in gm:
    if i != 'polityscore':
        gm[i] -= gm[i].mean()
```

```python
gm.mean()
```

```python
f = plt.figure(figsize=(10, 10))
plt.matshow(gm.corr(), fignum=f.number, cmap='bwr')
plt.xticks(range(gm.select_dtypes(['number']).shape[1]), gm.select_dtypes(['number']).columns, fontsize=14, rotation=90)
plt.yticks(range(gm.select_dtypes(['number']).shape[1]), gm.select_dtypes(['number']).columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
```

```python
for i in gm:
    if i != 'polityscore':
        plt.figure()
        sns.regplot(x=i, y='polityscore', data=gm)
```

```python
reg = smf.ols('polityscore ~ armedforcesrate + breastcancerper100th + alcconsumption', data=gm).fit()
print(reg.summary())
```

```python
#Q-Q plot for normality
fig = sm.qqplot(reg.resid, line='r')
```

```python
# simple plot of residuals
stdres = pd.DataFrame(reg.resid_pearson)
plt.plot(stdres, 'o', ls='None')
l = plt.axhline(y=0, color='r')
plt.ylabel('Standardized Residual')
plt.xlabel('Observation Number');
```

```python
# additional regression diagnostic plots
for i in reg.params.index:
    if i != 'Intercept':
        fig = plt.figure(figsize=(12,8))
        fig = sm.graphics.plot_regress_exog(reg,  i, fig=fig)
```

```python
# leverage plot
fig = sm.graphics.influence_plot(reg, size=9)
```
